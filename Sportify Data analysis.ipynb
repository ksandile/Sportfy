# Now we will put release date in place of datetime

```python
df_track.set_index("release_date", inplace=True)
df_track.index=pd.to_datetime(df_track.index)
df_track.head()
```

# Now we will check which artist is alocated in column 18 our dataset.

# Now we will convert our duration time from ms to seconds

```python
df_track["duration"] = df_track["duration_ms"].apply(lambda x: round(x/1000))
df_track.drop("duration_ms", inplace=True, axis=1)
```

```python
df_track.duration.head()
```

Now we can see that our duration values have been change from ms to seconds

# Now lets create a sample DataFrame, using the original dataframe.

```python
sample_df = df_track.sample(int(0.004*len(df_track)))
```

```python
print(len(sample_df))
```

# Now lets create a regretion plot between loudness and energy.

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy Correlation")
```

As we can see that there is a very high correlation between Loud and energy.
Cause all our songs are in one direction.
When the energy increases the loudness of the song also increases.

# We will also do the same in popularity and acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "b").set(title = "popularity vs acousticness Correlation")
```

When acousticness of the song increases, the popularity of the song decreases.
When popularity of the song increases, the acousticness decreases.

# Now we will create a column called year, from the column release date.

```python
df_track
```

```python
df_track['dates']=df_track.index.get_level_values('release_date')
df_track.dates=pd.to_datetime(df_track.dates)
years=df_track.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5 ,kind="hist").set(title="Number of songs per year")
```

So this plot tells us the number of songs has increase each year in the dataset, from the release year since music has became more excessable to people.

- As you can see in 1920 they were few, the as years went by it increases.
- And in 2020 we have a lot of songs that people can listen too.

# Now we will see a duration of songs over the years

```python
total_dr = df_track.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

# Now we will create a line plot to analyse duration of the songs by year.

```python
total_dr = df_track.duration
fig_dims = (10,5)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x = years, y = total_dr, ax = ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```

As you can see from our bar and line graph that song duration from 1920 was low, but they started to increase their duration.
Then after 2010 they start to decrease as they continue decreasing up to 2020

# Now lets move to second data analyses, which is based on Genre of the songs

```python
df_genre = pd.read_csv("SpotifyFeatures.csv")
```

```python
df_genre.head()
```

# Now we will see a duration of songs in different genre

```python
plt.title("Duration of songs in Different Genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y = 'genre', x = 'duration_ms', data = df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

Then we can swee that world genre has a high duration
And also Classical genre has a second high duration
Then we we have a lowest duration, which is Children's Music.

# Now we will want to see a top 5 genre by popularity

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize = (10,5))
famous = df_genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = famous).set(title = "Top 5 Genres by Popularity")
```

Those are our top 5 Genre by Popularity
